## Calculating Distance

For this assignment was are using additional distance measures so I have desided to create different functions to calculate the different similarities. For the assignement we were suppose to use Euclidean distance, Jaccard and Cosine Similarity for the distance, but the function provided for the Jaccard does not yeild results I have decided to go ahead and calculate the distances using Euclidean distance, Manhattan distance and Cosine Similarity.

```python
# choose your own way to calculate distance

from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import manhattan_distances
from scipy.stats.stats import pearsonr
from sklearn.metrics.pairwise import cosine_similarity
```

The first function is with the normal Euclidean distances measure.

```python

warnings.filterwarnings("ignore")

ALL_FEATURES = ['review_overall', 'review_aroma', 'review_palate', 'review_taste']
def calculate_similarity(beer1, beer2):
    # find common reviewers
    beer_1_reviewers = df[df.beer_name==beer1].review_profilename.unique()
    beer_2_reviewers = df[df.beer_name==beer2].review_profilename.unique()
    common_reviewers = set(beer_1_reviewers).intersection(beer_2_reviewers)

    # get reviews
    beer_1_reviews = get_beer_reviews(beer1, common_reviewers)
    beer_2_reviews = get_beer_reviews(beer2, common_reviewers)
    dists = []
    for f in ALL_FEATURES:
        dists.append(euclidean_distances([beer_1_reviews[f]], [beer_2_reviews[f]])[0][0])
    return dists

calculate_similarity(beer_1, beer_2)
```

The next function is with the new Manhattan distances measure.

```python
warnings.filterwarnings("ignore")

def calculate_similarity2(beer1, beer2):
    # find common reviewers
    beer_1_reviewers = df[df.beer_name==beer1].review_profilename.unique()
    beer_2_reviewers = df[df.beer_name==beer2].review_profilename.unique()
    common_reviewers = set(beer_1_reviewers).intersection(beer_2_reviewers)

    # get reviews
    beer_1_reviews = get_beer_reviews(beer1, common_reviewers)
    beer_2_reviews = get_beer_reviews(beer2, common_reviewers)
    dists2 = []
    for f2 in ALL_FEATURES:
        dists2.append(jaccard_similarity(beer_1_reviews[f2], beer_2_reviews[f2]))
    return dists2

calculate_similarity2(beer_1, beer_2)
```

And the last function is with the Cosine similarity measure.

```python
warnings.filterwarnings("ignore")

def calculate_similarity3(beer1, beer2):
    # find common reviewers
    beer_1_reviewers = df[df.beer_name==beer1].review_profilename.unique()
    beer_2_reviewers = df[df.beer_name==beer2].review_profilename.unique()
    common_reviewers = set(beer_1_reviewers).intersection(beer_2_reviewers)

    # get reviews
    beer_1_reviews = get_beer_reviews(beer1, common_reviewers)
    beer_2_reviews = get_beer_reviews(beer2, common_reviewers)
    dists3 = []
    for f3 in ALL_FEATURES:
        dists3.append(cosine_similarity([beer_1_reviews[f3]], [beer_2_reviews[f3]])[0][0])
    return dists3

calculate_similarity3(beer_1, beer_2)
```

## Allow the User to Customize the Weights

Now that all the different distances have been found I want to see what the effect is between them and what is the effect when changing the weights between them. To do this I am only going to look at the difference between the 90 Minute IPA and Old Raspution Russion Imperial Stout, being the most reviewed and the second most reveiwed beers. To examine the different weights I have decided to first look at equal weights, cosidered weighting of none, then when overall distance is weighted heavier than the rest, cosidered weighting 1, and finally when palate distance and taste distance is waited above the rest, considered weighting 2. It should be noted the first weight only has a sum of weights to be 4 and the other two have sum of weights to be 6.

```python
def calc_distance(dists, beer1, beer2, weights):
    mask = (dists.beer1==beer1) & (dists.beer2==beer2)
    row = dists[mask]
    row = row[['overall_dist', 'aroma_dist', 'palate_dist', 'taste_dist']]
    dist = weights * row
    return dist.sum(axis=1).tolist()[0]

weights = [1, 1, 1, 1]
weights1 = [3, 1, 1, 1]
weights2 = [1, 1, 2, 2]
#fixed from PY 2.7
print("For the difference between 90 Minute IPA & Old Rasputin Russian Imperial Stout:")
print("Dist:Euclidean Weight:None = %s" % calc_distance(simple_distances, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights))
print("Dist:Euclidean Weight:1 = %s" % calc_distance(simple_distances, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights1))
print("Dist:Euclidean Weight:2 = %s" % calc_distance(simple_distances, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights2))
print("Dist:Manhattan Weight:None = %s" % calc_distance(simple_distances2, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights))
print("Dist:Manhattan Weight:1 = %s" % calc_distance(simple_distances2, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights1))
print("Dist:Manhattan Weight:2 = %s" % calc_distance(simple_distances2, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights2))
print("Dist:Cosine Weight:None = %s" % calc_distance(simple_distances3, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights))
print("Dist:Cosine Weight:1 = %s" % calc_distance(simple_distances3, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights1))
print("Dist:Cosine Weight:2 = %s" % calc_distance(simple_distances3, '90 Minute IPA', 'Old Rasputin Russian Imperial Stout', weights2))
```

It's easy to see right off the bat that the difference between the distance measure types comes from the way the distance was calculated, so it's hard to see which change in weight had the biggest effect to different models because all the distance are so unique. It is easily to see that with the same weight all the way across for each distance that was the lowest overall total because there were less weights to those than to other two weights. But what I thought was really interesting is for the Euclidean and Manhattan the weighting of 1 distance was greater than the weighting of 2 distance, but that wasn't the case for the Cosine distance.
